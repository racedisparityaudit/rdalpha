```python
class MeasureSet(object):
    self.name = ''
    

class MeasureSetFactory(object):
    
    def __init__(self, data_categories = [], measure_categories = [], standard_categories = ['Race', 'Race_type', 'Location', 'Location_type', 'Income', 'Income_type']):
        self.standard_categories = standard_categories
        self.data_categories = []
        self.measure_categories = []
        self.all_categories = standard_categories + data_categories + measure_categories
        
        self.measures = []
    
    def import_measures_from_excel(self,excel_path):
        data = pd.read_excel(io=excel_path, sheetname="Data")
        self.full_data = self.data_with_extra_columns(data)
        
        self.measures = pd.read_excel(io=excel_path, sheetname="Measures").fillna('')
        
        unique_measures = measures['Measure'].unique()
        for measure_name in unique_measures:
            measure_data = self.read_all_measure_data(measure_name)
            print(measure_data)
        
    def read_all_measure_data(self,measure):
        final_columns = self.all_categories + [measure]
        df = pd.DataFrame(columns=final_columns)
        
        group = self.measures[measures['Measure'] == measure]
        for index, sub_measure in group.iterrows():
            all_columns = self.all_categories + [sub_measure['Code']]

            # Fill the measure columns with values from the category
            for measure_category in self.measure_categories:
                full_data[measure_category] = sub_measure[measure_category]

            # Get a subset of the data
            frame = full_data[all_columns]

            # Add it to our existing data
            frame.columns = final_columns
            df = df.append(frame)
        return df
        
        
    def data_with_extra_columns(self,data):
        full_data = data
        length = len(full_data['Race'])
        for mc in measure_categories:
            full_data[mc] = pd.Series(np.zeros(length))
        return full_data
```
