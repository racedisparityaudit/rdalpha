```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel(io="../data/input/datasets/attainment/ks4attainment_2016_18plus_gender.xlsx", sheetname="Data")
data.head(3)
```

```python
measures = pd.read_excel(io="../data/input/datasets/attainment/ks4attainment_2016_18plus_gender.xlsx", sheetname="Measures")
measures = measures.fillna('')
measures.head(3)
```

```python
unique_measures = measures['Measure'].unique()
unique_measures[0:3]
```

```python
group = measures[measures['Measure'] == 'Total Pupils']
group
```

```python
standard_cells = ['Race', 'Race_type', 'Location', 'Location_type', 'Income', 'Income_type']
data_categories = []
measure_categories = ['Gender','Measure']
```

```python
full_data = data
length = len(full_data['Race'])
for mc in measure_categories:
    full_data[mc] = pd.Series(np.zeros(length))
full_data.head(3)
```

```python
all_categories = standard_cells + data_categories + measure_categories
all_categories

```

```python
final_columns = all_categories + ['Total Pupils']
df = pd.DataFrame(columns=final_columns)

for index, sub_measure in group.iterrows():
    all_columns = all_categories + [sub_measure['Code']]
    
    # Fill the measure columns with values from the category
    for measure_category in measure_categories:
        full_data[measure_category] = sub_measure[measure_category]
    
    # Get a subset of the data
    frame = full_data[all_columns]
    
    # Add it to our existing data
    frame.columns = final_columns
    df = df.append(frame)

df
```

```python
serieses = pd.groupby(df, ['Race_level', 'Location_level'])

for series in serieses:
    print(series)
```
